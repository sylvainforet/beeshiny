```python
bee_paths = []
#long_tracked_bees_id_group = remove_short_tracking.groupby('BeeID',  sort=True)
grouped_bees_by_id = data.groupby('BeeID',  sort=True)

tag_coord_counts = {1: 0, 2: 0, 3: 0}

class BeePaths:
    __slots__ = ['bee_id', 'tag', 'paths_index', 'coords']
    def __init__(self, bee_id, tag, paths_index, coords):
        self.bee_id = bee_id
        self.paths_index = paths_index
        self.tag = tag
        self.coords = coords

min_frames_bee_visible = 250
tag_percent_confidence = 0.7
min_number_tags_classified = 100

for group_id, df in grouped_bees_by_id:
    #print (group_id)
    if len(df['Frame']) < min_frames_bee_visible:
        #print ("Skipping", len(df['Frame']))
        continue
    list_of_paths = []
    list_of_x_coords = []
    list_of_y_coords = []
    new = True
    max_frame = df['Frame'].max()
    prev_bee_index = 0
    remove_zeroes = df[df.Tag > 0]
    if len(remove_zeroes['Tag']) < min_number_tags_classified:
        ### Again, keep track of what you are filtering out
        continue
    tag_percentage = remove_zeroes['Tag'].value_counts().max() / len(remove_zeroes['Tag'])
    tag_identity = remove_zeroes['Tag'].value_counts().idxmax()
    if tag_percentage < tag_percent_confidence:
        ### Again, keep track of what you are filtering out
        continue
    for i, value in df['Frame'].iteritems():
        if new:
            start_path = i
            new = False
            prev_frame_index = i
            list_of_x_coords.append(data['X'][i])
            list_of_y_coords.append(data['Y'][i])
        else:
            difference_prev_frame = df['Frame'][i] - df['Frame'][prev_frame_index]
            if difference_prev_frame == 1:
                list_of_x_coords.append(data['X'][i])
                list_of_y_coords.append(data['Y'][i])
                prev_frame_index = i
                if value == max_frame:
                    #tag_coord_counts[tag_identity] += len(single_path)
                    list_of_paths.append((start_path, i))
            else:
                #tag_coord_counts[tag_identity] += len(single_path)
                list_of_paths.append((start_path, prev_frame_index))
                start_path = i
                prev_frame_index = i

    if len(list_of_paths) > 0:
        bee_paths.append(BeePaths(group_id, tag_identity, list_of_paths, np.array([list_of_x_coords, list_of_y_coords])))
        #print(len(list_of_paths))
        
print(len(bee_paths))
```
